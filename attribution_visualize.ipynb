```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with real CSV later)
data = {
    'source': ['google', 'facebook', 'linkedin', 'twitter', 'unknown'],
    'signups': [120, 90, 45, 30, 20]
}
df = pd.DataFrame(data)

# Plot bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(df['source'], df['signups'], color='skyblue')
plt.title('Attribution by Source')
plt.xlabel('Source')
plt.ylabel('Number of Signups')

# Annotate bars with numbers
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),
                 textcoords="offset points",
                 ha='center', va='bottom')

plt.tight_layout()
plt.savefig('outputs/attribution_chart.png')
plt.show()
```
